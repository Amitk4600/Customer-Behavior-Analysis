```python
import pandas as pd
```

```python
df=pd.read_csv("data/customer_shopping_behavior.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df =df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a columns age_group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

```python
# create a column purchase_frequancy_days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-weekly':14,
    'Annually':365,
    'Every 3 Months':90

}

df['purchase_frequancy_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequancy_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head()
```

```python
#check the 'discount_applied', 'promo_code_used' are same or not

(df['discount_applied']== df[ 'promo_code_used']).all()
```

```python
df = df.drop('promo_code_used',axis=1)
```
